## **Using Linear Regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("SLR_house_prices.csv")
data.head(10)
```

**Perform EDA**

```python
#shape
data.shape
```

```python
#info
display(data.info())
```

```python
# X ,y =[data['area'], data['price']]
```

```python
#describe

data.describe()
```

```python
#null value in both columns
data.isnull().sum()
```

```python
#col 1 area mean
area_mean=data.area.mean()
area_mean
```

```python
#col 2 price mean
price_mean=data.price.mean()
price_mean
```

```python
#fill both cols area & mean with his mean value
data.area=data.area.fillna(area_mean)
data.price=data.price.fillna(price_mean)
```

```python
# check again
data.isnull().sum()
```

```python
plt.scatter(data.area,data.price)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data.area,data.price,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
model.fit(x_train.values.reshape(-1,1),y_train)
```

```python
pred_y=model.predict(x_test.values.reshape(-1,1))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,pred_y)
```

```python
plt.plot(range(0,len(y_test)),y_test)
plt.plot(range(0,len(pred_y)),pred_y)
plt.show()
```

```python
plt.plot(range(0,10),y_test[:10])
plt.plot(range(0,10),pred_y[:10])
plt.show()
```
